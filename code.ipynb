# Tic Tac Toe using two person zero sum and alpha beta pruning

```python
player,opponent='X','O'

def isMovesLeft(board):
    for row in board:
        for item in row:
            if item=='_':
                return True
            
    return False

def evaluate(board):
    
    # row wise
    for i in range(3):
        if board[i][0]==board[i][1] and board[i][1]==board[i][2]:
            if board[i][0]==player:
                return 10
            elif board[i][0]==opponent:
                return -10
    
    # col wise
    for i in range(3):
        if board[0][i]==board[1][i] and board[1][i]==board[2][i]:
            if board[0][i]==player:
                return 10
            elif board[0][i]==opponent:
                return -10
    
    
    # diagonal wise
    if board[0][0]==board[1][1] and board[1][1]==board[2][2]:
        if board[0][0]==player:
            return 10
        elif board[0][0]==opponent:
            return -10
        
    if board[0][2]==board[1][1] and board[1][1]==board[2][0]:
        if board[0][2]==player:
            return 10
        elif board[0][2]==opponent:
            return -10
        

def miniMax(board,depth,isMaxPlayer,alpha,beta):
    score=evaluate(board)
    
    if score==10:
        return score
    
    if score==-10:
        return score
    
    if not isMovesLeft(board):
        return 0
    
    if isMaxPlayer:
        maxScore=-1000
        
        for i in range(3):
            for j in range(3):
                if board[i][j]!='_':
                    continue

                board[i][j]=player
                
                score=miniMax(board,depth+1,not isMaxPlayer,alpha,beta)
                maxScore=max(maxScore,score)
                alpha=max(alpha,maxScore)
                
                board[i][j]='_'
                
                if alpha>=beta:
                    break
                    
            if alpha>=beta:
                break
                
        return depth-maxScore
    else:
        minScore=1000
        
        for i in range(3):
            for j in range(3):
                if board[i][j]!='_':
                    continue
                    
                board[i][j]=opponent
                
                score=miniMax(board,depth+1,not isMaxPlayer,alpha,beta)
                minScore=min(minScore,score)
                beta=min(beta,minScore)
                
                board[i][j]='_'
                
                if alpha>=beta:
                    break
                    
            if alpha>=beta:
                break
                
        return minScore+depth

def findBestMove(board):
    
    bestScore=-1000
    row,col=-1,-1
    
    for i in range(3):
        for j in range(3):
            if board[i][j]!='_':
                continue
                
            board[i][j]=opponent
            
            score=miniMax(board,0,False,-1000,1000)
            
            board[i][j]='_'
            
            if score>bestScore:
                bestScore=score
                row=i
                col=j
                
    return row,col

def printBoard(board):
    for row in board:
        for item in row:
            print(item,end=' ')
        print('')
        
    print('')

def game():
    board=[['_','_','_'],['_','_','_'],['_','_','_']]
    
    while True:
        printBoard(board)
        move=int(input('Enter your move(cell No): '))
        
        move-=1
        row=move//3
        col=move%3
        
        if board[row][col]!='_':
            print('Invalid Move! Try Again.')
        else:
            board[row][col]=player
            
            printBoard(board)
            
            score=evaluate(board)
            if score==10:
                print('You won!')
                break
            
            if not isMovesLeft(board):
                print('Draw!')
                break
                
            oppMove=findBestMove(board)
            board[oppMove[0]][oppMove[1]]=opponent
            
            printBoard(board)
            
            score=evaluate(board)
            if score==-10:
                print('Opponent Won!')
                break
                
                
            if not isMovesLeft(board):
                print('Draw!')
                break

game()
```
